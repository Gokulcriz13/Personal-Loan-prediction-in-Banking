# Data Preparation

```python
df2 = df.loc[:,['Income','CCAvg','Mortgage','CreditCard','Personal Loan']]
df2
```

# Train and Test

```python
from sklearn.model_selection import train_test_split 
```

```python
x = df2.iloc[:,df2.columns!='Personal Loan'] 
y = df2.iloc[:,df2.columns=='Personal Loan'] 
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.8) #80% of test
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

# Algorithm 1 - Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model1 = RandomForestClassifier()
```

```python
model1.fit(xtrain,ytrain.values.ravel())
```

```python
predict_output1 = model1.predict(xtest)
print(predict_output1)
```

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
```

```python
acc1 = accuracy_score(predict_output1,ytest)
print('The accuracy score for RF:',acc1)
```

```python
cm1 = np.array(confusion_matrix(ytest, predict_output1, labels=[1,0]))
confusion1 = pd.DataFrame(cm1, index=['Personal Loan Eligibility', 'Personal Loan Non-Eligibility'],columns=['predicted Eligibility PL','predicted Non-Eligibility PL'])
confusion1
```

```python
sns.heatmap(confusion1, annot=True)
```

```python
cr1 = classification_report(ytest,predict_output1)
print(cr1)
```
